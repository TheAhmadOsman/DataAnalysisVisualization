## Bubble Plot

Let's find the peak time during workdays for bike riders.

```python
workdayAvgByHr = bikeDays[bikeDays.workingday==1].groupby("hr").mean()["cnt"]
```

```python
workdayAvgByHr
```

```python
workdayAvgByHr.plot(kind="barh", width=0.8, figsize=(10,10))
```

```python
workDayBySeasonHour = bikeDays[bikeDays.workingday==1].groupby(["season", "hr"]).mean()
```

```python
workDayBySeasonHour
```

```python
workDayBySeasonHour.reset_index(level=0, inplace=True)
```

```python
workDayBySeasonHour
```

```python
workDayBySeasonHour.reset_index(level=0, inplace=True)
```

```python
workDayBySeasonHour
```

```python
workDayBySeasonHour.plot.scatter(x="season", y="hr", s=workDayBySeasonHour["cnt"])
```
